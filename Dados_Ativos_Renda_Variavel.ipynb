# Notebook com as funções necessárias para obter as informações de ativos de renda variável negociados na Bolsa de Valores

## Definição de imports

```python
# imports de bibliotecas
import pandas as pd
import yfinance as yf
import pdb
```

## Função para obter os retornos dos ativos negociados na bolsa de valores brasileira

```python
# Os dados obtidos são a partir de 01/01/1995
def buscarRetornosAtivosNegociadosBolsaBrasileira(array_tipoativos, data_inicio, data_fim):
  array_df_tipoativos = []
  array_df_tipoativos.append(array_tipoativos[0])
  array_df_tipoativos.append(array_tipoativos[1])
   # O "len(df_tipo_ativos)-1" serve para que o df de Renda Fixa não seja considerado
  colecao_retornos_calculados = []
  for i in range(len(array_df_tipoativos)):
    #pdb.set_trace()
    tickers = array_df_tipoativos[i].iloc[:,0]
    tickers.to_numpy()
    # Isso foi necessário para que o 'yf' recupere as informações dos ativos do Brasil
    tickers_SA = [x + '.SA' for x in tickers]
    precos_ativos = yf.download(' '.join(tickers_SA),data_inicio, data_fim )['Adj Close'].resample('Y').last()
    precos_ativos = precos_ativos.round(2)
    retornos_calculados = precos_ativos.pct_change()[1:]
    #retornos_calculados = retornos_calculados.round(2)

    if len(tickers_SA) == 1:
      retornos_calculados = pd.DataFrame(retornos_calculados)
      retornos_calculados.columns = tickers_SA

    colecao_retornos_calculados.append(retornos_calculados.round(2))
  return colecao_retornos_calculados
```

## Função para obter retornos de renda variável

```python
# Os dados obtidos são a partir de 01/01/1995
def buscarRetornosAtivosRendaVariavel(array_tipoativos, data_inicio, data_fim):
  colecao_retornos_calculados = []

  tickers = array_tipoativos

  tickers_SA = [x + '.SA' for x in tickers]
  precos_ativos = yf.download(' '.join(tickers_SA),data_inicio, data_fim )['Adj Close'].resample('Y').last()
  precos_ativos = precos_ativos.round(2)
  retornos_calculados = precos_ativos.pct_change()[1:]


  if len(tickers_SA) == 1:
    retornos_calculados = pd.DataFrame(retornos_calculados)
    retornos_calculados.columns = tickers_SA

  colecao_retornos_calculados.append(retornos_calculados.round(2))

  return colecao_retornos_calculados
```
