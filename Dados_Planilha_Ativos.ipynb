## Função para obter dados da planilha de ativos

```python
def RecuperarDadosPlanilhaAtivos():
  uploaded = files.upload()
  df_array = []

  xlsxFile = pd.ExcelFile('Planilha_Ativos.xlsx')
  sheets = xlsxFile.sheet_names

  for sheet in sheets:
    print(sheet)
    df = pd.read_excel('Planilha_Ativos.xlsx', sheet_name=sheet)
    df_array.append(df)

  return df_array
```
